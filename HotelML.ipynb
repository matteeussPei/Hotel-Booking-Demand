```python
df.query('children.isnull()')
```

```python
df = df.dropna(subset=['children', 'country']).reset_index(drop=True)
```

```python
df['total_days'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights']
df.drop(columns = ['stays_in_week_nights', 'stays_in_weekend_nights'], inplace = True)
df.head()
```

```python
filtro = (df['babies'] == 0) & (df['children'] == 0) & (df['adults'] == 0)
df = df[~filtro]
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop(columns=['country', 'reservation_status_date', 'reservation_status'], inplace = True)
```

```python
for i in range(df.shape[0]):
    if (df['assigned_room_type'][i] == df['reserved_room_type']):
        df['room_type'][i] = 0
    else:
        df['room_type'][i] = 1
```
